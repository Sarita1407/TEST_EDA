# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python
# Insert your code here...
# checking car column what are all brands are available other than ['Volkswagen','BMW','Mercedes-Benz','Ford','Nissan','Toyota','Renault']
carskeep=['Volkswagen','BMW','Mercedes-Benz','Ford','Nissan','Toyota','Renault']
data[data.car.isin(carskeep)==0][:50]
```

```python

#checking for car whise price value is zero
data[data['price']==0]
data.price.median()
# checking median  price for each car body
data.groupby(['body'])['price'].median()

#updating zero price value to median for each car each car body
data['price'] = np.where((data['price']==0) & (data['body']=='crossover') ,19400.0, data['price'])
data['price'] = np.where((data['price']==0) & (data['body']=='hatch') ,7000.0, data['price'])
data['price'] = np.where((data['price']==0) & (data['body']=='other') ,10575.0, data['price'])
data['price'] = np.where((data['price']==0) & (data['body']=='sedan') ,7800.0, data['price'])
data['price'] = np.where((data['price']==0) & (data['body']=='vagon') ,7200.0, data['price'])
data['price'] = np.where((data['price']==0) & (data['body']=='van') ,7999.0, data['price'])

#checking for car whose price value is zero again
data[data['price']==0]
```

```python
# car body column the few fields are with  other which is not valid cleaned by maximum value sedan
# others count is 838
data.body.value_counts()
# replace othere to sedan
data.body=data.body.astype(str).str.replace('other','sedan')
#checking count again 
data.body.value_counts()
```

```python
#change engine type other to hybrid or electric
data.engType.value_counts()
#replace engType column value other to Electric_or_Hybrid
data.engType= data.engType.astype(str).str.replace('Other', 'Electric_or_Hybrid')
#checking th count again
data.engType.value_counts()
```

```python
#car milage handling zero data and outliers data
data.mileage.value_counts()
# milage 0 value for 348 rows
data[data['mileage']==0]
# medain is 128
data['mileage'].median()
# mileage for each car body type
data.groupby(['body'])['mileage'].median()
# outlier data 2 rows
data[data['mileage']==999]

#median excluding  0 to median 128.0
data['mileage'] = np.where(((data['mileage']==0) | (data['mileage']==999)) & (data['body']=='crossover') , 112.0, data['mileage'])
data['mileage'] = np.where(((data['mileage']==0)| (data['mileage']==999)) & (data['body']=='hatch') , 100.0, data['mileage'])
data['mileage'] = np.where(((data['mileage']==0)| (data['mileage']==999)) & (data['body']=='sedan') , 131.0, data['mileage'])
data['mileage'] = np.where(((data['mileage']==0)| (data['mileage']==999)) & (data['body']=='vagon') , 175.0, data['mileage'])
data['mileage'] = np.where(((data['mileage']==0)|(data['mileage']==999)) & (data['body']=='van'),181.0,data['mileage'])

#car milage handling outliers data recheck
data[data['mileage']==999]
```

```python
#engV column value 99.99 outlier where 95-th percentile is 4.6
data.engV.value_counts()
data.engV.median()

data[data['engV']==99.99]

#median value engv for each engtype 
data.groupby(['engType'])['engV'].median()
#updating the engv vlaue to median value of each engtype
data['engV'] = np.where((data['engV']==99.99) & (data['engType']=='Diesel') ,2.14, data['engV'])
data['engV'] = np.where((data['engV']==99.99) & (data['engType']=='Electric_or_Hybrid') ,1.90, data['engV'])
data['engV'] = np.where((data['engV']==99.99) & (data['engType']=='Gas') ,2.00, data['engV'])
data['engV'] = np.where((data['engV']==99.99) & (data['engType']=='Petrol') ,1.80, data['engV'])


#checking the roes again 
data[data['engV']==99.99]
```
